```python
import pandas as pd
import numpy as np
data=np.random.rand(50,3)
dataframe=pd.DataFrame(data,columns=['c1','c2','c3'])
num_nulls = int(0.1 * dataframe.size)
numindexes=np.random.choice(dataframe.size,num_nulls,replace=False)
dataframe.values.flat[numindexes]=np.nan
print(dataframe)
print()
sum=dataframe.isnull().sum()
print("sum is")
print(sum)
print(sum.sum())
dataframe.dropna(thresh=len(dataframe)-5,axis=1)
print(dataframe.sum(axis=1))
maxlabel=dataframe.sum(axis=1).idxmax()
dataframe.drop(maxlabel)

dataframe.sort_values(by='c1')
dataframe['c1']=dataframe['c1'].drop_duplicates()
print(dataframe)
dataframe['c1'].corr(dataframe['c3'])
dataframe['c1'].cov(dataframe['c3'])

# g. Discretize the second column and create 5 bins
# print("(g)")
# print()

# bins=df["Column2_bins"] = pd.cut(df["Column2"], 5)

# print(bins)
# print()
# print("Final DataFrame : ")
# print(df)
```
